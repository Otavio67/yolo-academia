# Aula 11: Colocando seu Modelo em Produção (Inferência) 🚀


## 1. Copiando o modelo `best.pt`


```python
import os
import shutil

shutil.copy('runs/detect/train-2/weights/best.pt', 'modelo_academia.pt')

if os.path.exists('modelo_academia.pt'):
    print("Modelo copiado com sucesso! ✅")
else:
    print("Arquivo não encontrado. Verifique o caminho. ❌")
```

## 2. Carregando o Modelo Específico


```python
from ultralytics import YOLO

# Em vez de 'yolov8n.pt', carregamos O NOSSO ARQUIVO!
nome_modelo = 'modelo_academia.pt'

print(f"Carregando modelo: {nome_modelo}...")
model = YOLO(nome_modelo)

# Verificar quais classes esse modelo conhece
print("\nClasses que este modelo conhece:", model.names)
```

## 3. Realizando a Inferência


```python
import cv2

source = 'academia.mp4'

if not os.path.exists(source) and type(source) is str:
    print("Aviso: Vídeo local não encontrado, tente novamente...")

results = model.predict(source=source, save=True, conf=0.5)
```
