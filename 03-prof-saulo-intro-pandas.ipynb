```python
# Importar a biblioteca pandas
import pandas as pd

print("Pandas versão:", pd.__version__)
```

## 1. Criando um DataFrame a partir de um Dicionário

A forma mais simples e intuitiva de criar um DataFrame é usar um dicionário Python. As chaves do dicionário se tornam os nomes das colunas, e os valores (listas) se tornam os dados das colunas.

```python
# Criando um dicionário com dados de estudantes
dados_estudantes = {
    'nome': ['Alice', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'idade': [20, 22, 19, 21, 23],
    'nota': [8.5, 9.0, 7.8, 8.9, 9.2],
    'turma': ['A', 'B', 'A', 'C', 'B']
}

# Criando um DataFrame a partir do dicionário
df_estudantes = pd.DataFrame(dados_estudantes)

# Exibindo o DataFrame
print("DataFrame de Estudantes:")
print(df_estudantes)
```

### Explorando a estrutura do DataFrame

Vamos conhecer melhor nosso DataFrame usando métodos úteis:

```python
# Informações sobre o DataFrame
print("=== INFORMAÇÕES GERAIS ===")
print(f"Dimensões (linhas, colunas): {df_estudantes.shape}")
print(f"\nNomes das colunas: {df_estudantes.columns.tolist()}")
print(f"\nTipos de dados:\n{df_estudantes.dtypes}")
print(f"\nPrimeiras 2 linhas:\n{df_estudantes.head(2)}")
print(f"\nÚltimas 2 linhas:\n{df_estudantes.tail(2)}")
```

### Acessando dados do DataFrame

Existem várias formas de acessar dados em um DataFrame:

```python
# Acessando uma coluna completa
print("=== ACESSANDO COLUNAS ===")
print("Nomes dos estudantes:")
print(df_estudantes['nome'])
print(f"\nTipo: {type(df_estudantes['nome'])}")

# Acessando múltiplas colunas
print("\n=== MÚLTIPLAS COLUNAS ===")
print(df_estudantes[['nome', 'nota']])
```

```python
# Acessando linhas por posição (iloc)
print("=== ACESSANDO LINHAS ===")
print("Primeira linha (índice 0):")
print(df_estudantes.iloc[0])
print("\nTrês primeiras linhas:")
print(df_estudantes.iloc[0:3])
```

### Filtrando dados

Podemos filtrar o DataFrame usando condições:

```python
# Filtrando: Estudantes com nota >= 9.0
print("=== FILTRANDO DADOS ===")
estudantes_nota_alta = df_estudantes[df_estudantes['nota'] >= 9.0]
print("Estudantes com nota >= 9.0:")
print(estudantes_nota_alta)

# Filtrando: Estudantes de uma turma específica
print("\n\nEstudantes da turma 'A':")
turma_a = df_estudantes[df_estudantes['turma'] == 'A']
print(turma_a)
```

### Estatísticas básicas

Pandas oferece funções para calcular estatísticas dos dados:

```python
# Estatísticas descritivas
print("=== ESTATÍSTICAS DESCRITIVAS ===")
print(df_estudantes.describe())

# Funções específicas
print("\n\n=== FUNÇÕES ESPECÍFICAS ===")
print(f"Idade média: {df_estudantes['idade'].mean():.1f}")
print(f"Nota máxima: {df_estudantes['nota'].max()}")
print(f"Nota mínima: {df_estudantes['nota'].min()}")
print(f"Idade total: {df_estudantes['idade'].sum()}")
```

## 2. Importando dados de um arquivo CSV

CSV (Comma-Separated Values) é um formato comum para armazenar dados tabulares. Pandas facilita a importação de arquivos CSV com a função `read_csv()`.

Primeiro, vamos criar um arquivo CSV de exemplo para demonstrar:

```python
# Criando um arquivo CSV de exemplo
csv_conteudo = """produto,preco,quantidade,categoria
Notebook,3000.00,5,Eletrônicos
Mouse,50.00,20,Periféricos
Teclado,150.00,15,Periféricos
Monitor,800.00,8,Eletrônicos
Cadeira,500.00,10,Móveis
Mesa,1200.00,6,Móveis
Lámpada,80.00,25,Iluminação
"""

# Salvando o arquivo
with open('/home/saulo/projetos/visao/aula-03/src/produtos.csv', 'w') as f:
    f.write(csv_conteudo)

print("Arquivo CSV criado: /home/saulo/projetos/visao/aula-03/src/produtos.csv")
```

```python
# Importando dados do arquivo CSV
print("=== IMPORTANDO DADOS DO CSV ===\n")
df_produtos = pd.read_csv('/home/saulo/projetos/visao/aula-03/src/produtos.csv')

print("DataFrame carregado:")
print(df_produtos)
print(f"\n\nDimensões: {df_produtos.shape}")
print(f"\nTipos de dados:\n{df_produtos.dtypes}")
```

### Explorando os dados do CSV

Vamos explorar o DataFrame recém-carregado:

```python
# Informações sobre produtos
print("=== INFORMAÇÕES SOBRE PRODUTOS ===\n")
print("Primeiras 3 linhas:")
print(df_produtos.head(3))

print("\n\nProdutos por categoria:")
print(df_produtos['categoria'].value_counts())

print("\n\nPreço médio dos produtos:")
print(f"R$ {df_produtos['preco'].mean():.2f}")
```

### Operações com dados importados

Podemos realizar diversas operações com os dados:

```python
# Criando novas colunas (valor total em estoque)
df_produtos['valor_total'] = df_produtos['preco'] * df_produtos['quantidade']

print("=== PRODUTOS COM VALOR TOTAL ===\n")
print(df_produtos[['produto', 'preco', 'quantidade', 'valor_total']])

print(f"\n\nValor total em estoque: R$ {df_produtos['valor_total'].sum():.2f}")
```

```python
# Filtrando e ordenando
print("=== PRODUTOS CAROS (PREÇO > R$ 500) ===\n")
produtos_caros = df_produtos[df_produtos['preco'] > 500].sort_values('preco', ascending=False)
print(produtos_caros[['produto', 'preco', 'categoria']])

print("\n\n=== PRODUTOS COM MAIOR QUANTIDADE ===\n")
top_quantidade = df_produtos.nlargest(3, 'quantidade')[['produto', 'quantidade', 'categoria']]
print(top_quantidade)
```

## 3. Resumo de Conceitos Principais

### Funções úteis aprendidas:
- **`pd.DataFrame()`**: Criar um DataFrame a partir de um dicionário
- **`pd.read_csv()`**: Importar dados de um arquivo CSV
- **`.head()`** e **`.tail()`**: Ver primeiras e últimas linhas
- **`.describe()`**: Obter estatísticas descritivas
- **`.shape`**: Dimensões do DataFrame
- **`.columns`**: Nomes das colunas
- **`.dtypes`**: Tipos de dados das colunas
- **Filtragem**: `df[df['coluna'] > valor]`
- **Seleção de colunas**: `df[['col1', 'col2']]`
- **Operações**: Criar novas colunas, ordenar, filtrar
